```python
# Import OpenAI API
import openai
```

```python
# Load OpenAI API key
with open("openai.key", "r") as file:
    openai.api_key = file.read().strip()
```

```python
# Create a completion
response = openai.Completion.create(
  model="text-davinci-003",
  prompt="Correct this to standard English:\n\nShe no went to the market.",
  temperature=0,
  max_tokens=60,
  top_p=1.0,
  frequency_penalty=0.0,
  presence_penalty=0.0
)
```

```python
# Print the completion
print(response["choices"][0]["text"])
```

```python
def grammar_correction(sentence):
    # Create a completion
    response = openai.Completion.create(
      model="text-davinci-003",
      prompt=f"Correct this to standard English:\n\n{sentence}",
      temperature=0,
      max_tokens=60,
      top_p=1.0,
      frequency_penalty=0.0,
      presence_penalty=0.0
    )

    # Return the first choice of the completion
    return response["choices"][0]["text"]
```

```python
sentence = "How old is you? Prof.Shanfengfeng"
print(grammar_correction(sentence))
```
